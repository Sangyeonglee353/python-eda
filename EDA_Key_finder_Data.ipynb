```python
## 최종 데이터 형태
# result = [{
#     "Function Name" : "",
#     "Hot Key" : "",
#     "Category": "",
#     "Tool Name": "",
# },...]

resultKeyList = []
for i in range(len(data["results"])):
  # 1. 사용 툴
  toolName = data["results"][i]["properties"]["Tool Name"]["multi_select"][0]["name"]

  # 2. 카테고리
  category = data["results"][i]["properties"]["Category"]["multi_select"][0]["name"]

  # 3. 단축키
  length = len(data["results"][i]["properties"]["Hot key"]["rich_text"])
  hotKeyList = [];

  for j in range(length):
    value = data["results"][i]["properties"]["Hot key"]["rich_text"][j]["text"]["content"]
    hotKeyList.append(value)

  hotkey = ''.join(hotKeyList);

  # 4. 기능명
  functionName = data["results"][i]["properties"]["Function Name"]["title"][0]["plain_text"]

  # 5. 각 속성 설정
  resultKey = {
      "functionName" : "",
      "hotkey" : "",
      "category" : "",
      "toolName" : "",
  }

  resultKey["functionName"] = functionName
  resultKey["hotkey"] = hotkey
  resultKey["category"] = category
  resultKey["toolName"] = toolName

  # 6. 전체 배열에 추가
  resultKeyList.append(resultKey)

print(resultKeyList)
```

# 데이터 형태 가공 List to Object

```python
# 형식 변환
resultObject = {"data" : []}
resultObject["data"] = resultKeyList

print(resultObject)
```

```python
# JSON 파일 형태로 저장
import json

# Google Drive에 저장할 파일 경로 설정
file_path = '/content/drive/My Drive/data/convertNotionDB.json'

with open(file_path, 'w', encoding='utf-8') as json_file:
  json.dump(resultObject, json_file, ensure_ascii=False, indent=4)
```
